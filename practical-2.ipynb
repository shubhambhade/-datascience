```python
df2=dataframe5.drop(columns=['Name'])
df2
```

```python
def DetectOutlier(df2,var): 
 # IQR method is used to deal with outliers
 Q1 = df2[var].quantile(0.25)
 Q3 = df2[var].quantile(0.75)
 IQR = Q3 - Q1
 high, low = Q3+1.5*IQR, Q1-1.5*IQR
 
 print("Highest allowed in variable:", var, high)
 print("lowest allowed in variable:", var, low)
 count = df2[(df[var] > high) | (df2[var] < low)][var].count()
 print('Total outliers in:',var,':',count)
 # new dataframe is created which contains outliers
 df3 = df2[((df[var] < low) | (df2[var] > high))] #these are outliers
 print('Outliers : \n', len(df3))
 print(df3.T)
 df2 = df2[((df2[var] >= low) & (df2[var] <= high))] #now filter out data which is not outlier
 return(df2)
```

```python
# Outliers can be visualized using boxplot
# using seaborn library we can plot the boxplot
fig, axes = plt.subplots(2,2)
fig.suptitle('Before removing Outliers')
sns.boxplot(data = df2, y='Maths', ax=axes[0,0])
sns.boxplot(data = df2, y ='English', ax=axes[0,1])
sns.boxplot(data = df2, y ='Science', ax=axes[1,0])
sns.boxplot(data = df2, y ='History', ax=axes[1,1])
plt.show()
```

```python
# Remove outliers and display boxplots after removing outliers
df_cleaned = DetectOutlier(df2, 'Maths')

sns.boxplot(data=df_cleaned, y='Maths')
```
