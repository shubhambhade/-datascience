```python
fig, axes = plt.subplots(2,2,figsize = (16,9))
sns.histplot(df['sepal_length'], ax = axes[0,0])
sns.histplot(df['sepal_width'], ax = axes[0,1])
sns.histplot(df['petal_length'], ax = axes[1,0])
sns.histplot(df['petal_width'], ax = axes[1,1])
```

```python
print('boxplot of petal_length')
sns.boxplot(df['petal_length'])
```

```python
print('boxplot of petal_width')
sns.boxplot(df['petal_width'])
```

```python
print('boxplot of sepal_length')
sns.boxplot(df['sepal_length'])
```

```python
print('boxplot of sepal_width')
sns.boxplot(df['sepal_width'])
```

```python
fig, axes = plt.subplots(2,2,figsize=(16,9))
sns.boxplot(x='species',y='petal_length',data = df, ax=axes[0,0])
sns.boxplot(x='species',y='petal_width',data = df, ax=axes[0,1])
sns.boxplot(x='species',y='sepal_length',data = df, ax=axes[1,0])
sns.boxplot(x='species',y='sepal_width',data = df, ax=axes[1,1])
```
